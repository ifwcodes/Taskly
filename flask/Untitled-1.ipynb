```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "Hello! My app is running without the terminal!"


if __name__ == '__main__':
    app.run(debug=True, host='0.0.0.0', port=5000)
```

```python
todos = ["Buy groceries", "Read a book", "Walk the dog"]

@app.route('/todos')
def todo_list():
    return "<br>".join(todos)
```

```python
import os

os.makedirs('static', exist_ok=True)
app.static_folder = 'static'
```

```python
import os
from flask import Flask

if 'app' not in globals():
    app = Flask(__name__)

todos = ["Buy groceries", "Read a book", "Walk the dog"]

if '/todos' not in {rule.rule for rule in app.url_map.iter_rules()}:
    @app.route('/todos')
    def todo_list():
        return "<br>".join(todos)

os.makedirs('static', exist_ok=True)
app.static_folder = 'static'
```
